```python
import pickle as plk
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import normalize
from sklearn.utils import shuffle
```

```python
test_file = open("./data/test.pkl","rb")
train_file = open("./data/train.pkl","rb")
```

```python
test_data = plk.load(test_file)
train_data = plk.load(train_file)
```

```python
X_train = np.array([x[0] for x in train_data]).reshape(-1,1)
Y_train = np.array([x[1] for x in train_data])

x_test = np.array([x[0] for x in test_data] ).reshape(-1,1)
y_test = np.array([x[1] for x in test_data] )

```

```python
plt.figure(figsize=(16,8))

plt.xlabel("House Area")
plt.ylabel("House value")

plt.scatter(X_train,Y_train,s=1.5,label="Training set")
plt.scatter(x_test,y_test,label="Testing Set")

plt.legend()

plt.show()
```

# Task 2.2.2

### Shuffling and splitting the test and train data into 16 sets

```python
#shuffling
X_train,Y_train = shuffle(X_train,Y_train,random_state=42)

#splitting
x_train_datas = np.array(np.split(X_train,16))
y_train_datas = np.array(np.split(Y_train,16))
```

### Creating and Training the linear-regression models on the above obtained datas for polynomial of degree 1 to 15 

```python
all_models = {}
for power in range(1,16):
    #iterating through degrees
    model_array = []
    for x_data , y_data in zip(x_train_datas,y_train_datas):
        # tansforming data into curve of given degree
        poly = PolynomialFeatures(power)
        transformed_data = poly.fit_transform(x_data)
        
        #defining the model
        current_model = LinearRegression()
        
        #fitting the transformed 
        current_model.fit(transformed_data,y_data)
        model_array.append(current_model)
        
    all_models[power] = model_array
```

### Predicting the values of $Y_{hat}$ for each model   

```python
y_hats_pow = []
for power,models in all_models.items():
    y_hats = []
    
    poly = PolynomialFeatures(power)
    transformed_data = poly.fit_transform(x_test)
    
    for model in models:
        y_hats.append(model.predict(transformed_data))
    
    y_hats_pow.append(y_hats)
y_hats_pow = np.array(y_hats_pow)
```

## Plotting the predicted values

```python
power = 3
model_no = 0

y_hats = y_hats_pow[power]
y_hat = y_hats[model_no]

plt.figure(figsize=(16,8))

plt.scatter(x_test,y_test,label="Test Data")
plt.scatter(X_train,Y_train,label="Train Data",s=1.5)
x_test,y_hat = zip(*sorted(zip(x_test,y_hat)))

plt.plot(x_test,y_hat,label="Fitted curve")

plt.legend()
plt.show()

x_test = np.array([x[0] for x in test_data] ).reshape(-1,1)
y_test = np.array([x[1] for x in test_data] )
```

### Calculating the value of mean across the models for all powers

```python
mean_arr = []
for y_hats in y_hats_pow:
    exp_y_hat = np.mean(y_hats,axis=0)
    mean_arr.append(exp_y_hat)
mean_arr = np.array(mean_arr)
```

### Calculating average bias for all the models fitting all powers

The average bias is given as 

$$ Bias = E[\hat{f}(x)] - f(x) $$
where $E$ is expected value, $\hat{f}(x)$ is predicted value and $f(x)$ is label

```python
avg_biases_squared = np.square(mean_arr-y_test).mean(axis=1)

df = pd.DataFrame(avg_biases_squared)
df.columns = ['Average bias squared']
df['Power of fitted curve'] = df.index+1

df
```

### Plotting the bias w.r.t. power of the curve fitted 

```python
plt.figure(figsize=(16,8))
plt.plot(range(1,16),avg_biases_squared,label="Bias square")
plt.legend()
plt.show()
```

# Variance

The variance is given as
$$Variance = E[( \hat{f}(x) - E[\hat{f}(x)] )^2 ]$$

Where, $E$ is expected value and $\hat{f}(x)$ is the predicted value 

```python
avg_vars = []
for y_hats_list,avg in zip(y_hats_pow,mean_arr):
    var = ((y_hats_list-avg)**2).mean(axis=1)
    avg_vars.append(var.mean())

    
df = pd.DataFrame(avg_vars)
df.columns = ['Average Variance']
df['Power of fitted curve'] = df.index+1

df
```

```python
plt.figure(figsize=(16,8))


plt.plot(range(1,16),avg_vars,label="Average variance")
plt.plot(range(1,16),avg_biases_squared,label="Average bias")

plt.legend()
plt.xlabel("Degree of the curve fitted")
plt.ylabel("Bias / Variance")

plt.show()
```

# Task 2.3

## Calculating Irreducable Error

### 1) In order to calculate Irreducable Error we have to first calculate MSE
$$MSE = MEAN[(f(x) - \hat{f}(x))^2]$$

```python
MSEs = np.square((y_hats_pow-y_test)).mean(axis=2).mean(axis=1)
```

### 2) now calculating Irreducable Error
$$σ^2 = MSE − (Bias^2 + Variance)$$ \
where $f(x)$ represents the true value, $f'(x)$ represents the predicted value,
$MSE$ is the mean squared error and $σ^2$ represents irreducible error.

```python
IE = MSEs - (avg_biases_squared + avg_vars)
```

```python
plt.figure(figsize=(16,8))


plt.plot(range(1,16),avg_vars,label="Average variance")
plt.plot(range(1,16),avg_biases_squared,label="Average bias")
plt.plot(range(1,16),MSEs,label="MSE")
plt.plot(range(1,16),IE,label="Irreducable Error")

plt.legend()
plt.xlabel("Degree of the curve fitted")
plt.ylabel("Bias / Variance")

plt.show()
```
